```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
```

```python
X, _ = make_moons(n_samples=300, noise=0.05, random_state=42)
```

```python
# 2. Standarisasi fitur
X_scaled = StandardScaler().fit_transform(X)
```

```python
# 3. Terapkan DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=5)
labels = dbscan.fit_predict(X_scaled)
```

```python

# 4. Plot hasil cluster
plt.figure(figsize=(7, 5))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap="plasma", s=40, edgecolor="k")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Hasil Clustering DBSCAN pada Dataset Moons")
plt.show()
```

```python
# dataframe untuk lihat hasil
df = pd.DataFrame(X_scaled, columns=["Feature1", "Feature2"])
df["Cluster"] = labels
df
```
